# Test network config

```python
input_imgs = train['image']
input_labels = train['label']
```

```python
unet_kwargs = {
    'model_channels': 64,
    'channel_mult': [1, 2, 4],
    'num_blocks': 3,
    'attn_resolutions': [14, 7],
}

test_net = edm2_net.PrecondFlowMap(
    img_resolution=28,
    img_channels=1,
    label_dim=10,
    sigma_data=0.5,
    logvar_channels=64,
    unet_kwargs=unet_kwargs
)
```

```python
prng_key = jax.random.PRNGKey(42)

ex_s = ex_t = np.zeros((1,))
ex_img = input_imgs[0]
ex_img = ex_img.reshape((1, *ex_img.shape))
ex_label = jax.nn.one_hot(input_labels[0], num_classes=11).reshape((1, -1))

print('img shape', ex_img.shape)
print('t shape', ex_s.shape)
print('s shape', ex_s.shape)
print('label shape', ex_label.shape)

variables = test_net.init(prng_key, ex_s, ex_t, ex_img, ex_label, train=False, calc_weights=True)
print(f"Number of parameters: {ravel_pytree(variables)[0].size}")
```

```python
test_net.apply(variables, ex_s, ex_t, ex_img, ex_label, train=False)
```

```python
output1 = test_net.apply(variables, ex_s, ex_t, ex_img, ex_label, train=False, rngs={'params': prng_key})
```

```python
prng_key = jax.random.split(prng_key)[0]
output2 = test_net.apply(variables, ex_s, ex_t, ex_img, ex_label, train=False, rngs={'params': prng_key})
```

```python
np.allclose(output1, output2)
```
